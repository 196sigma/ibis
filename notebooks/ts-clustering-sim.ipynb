## Hamming Distance

The Hamming distance could be particularly effective here. It measures the number of positions at which the corresponding elements are different. Since the time series are binary, the Hamming distance directly gives you the count of mismatches between two series, making it a straightforward and interpretable measure of similarity.

```python
Y
```

```python
def hamming_distance(x, y):
    """Compute the Hamming distance between two binary sequences"""    
    return sum(el1 != el2 for el1, el2 in zip(x, y))

target_day = "2020-12-31"
res = {}
day = days[0]
print(day)
target_series = Y.query(f"day == '{target_day}'").iloc[:, 2]
for day in days:
    current_series = Y.query(f"day == '{day}'").iloc[:, 2]
    distance = hamming_distance(target_series, current_series)
    res[day] = distance
distance_df = pd.DataFrame(res, index=["distance"]).T.reset_index().rename(columns={"index": "day"})
distance_df.sort_values("distance", ascending=True, inplace=True)
distance_df

```

## Jaccard Similarity:

Another relevant measure could be the Jaccard similarity, which is defined as the size of the intersection divided by the size of the union of the sample sets. For binary sequences, this can indicate how similar the occurrences of 1s are between two sequences.

## Pearson Correlation:

If you're interested in how similarly two time series move together (i.e., when one increases or decreases, so does the other, regardless of the absolute values), then Pearson correlation is a good measure. It assesses the linear relationship between series.

## Euclidean Distance or Manhattan Distance:

For comparing the overall differences in the magnitudes of two time series, straightforward measures like Euclidean or Manhattan distance can be useful. They are especially relevant if the series are aligned in time and of the same length.

## Cosine Similarity

If the direction of the changes in the time series matters more than the magnitude of the changes, cosine similarity can be effective. It measures the cosine of the angle between two vectors (in this case, time series), providing insight into how similarly they move without being affected by the scale of the movements.
